```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16) #512x512 hepsi sıfır old için full siyah
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(384,10),pt2=(500,150),color=(255,0,0),thickness=10) #pt1 sol üst, pt2 sağ alt, color rgb, thickness pixel
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(0,255,0),thickness=8)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(400,400),color=(255,0,0),radius=30,thickness=-1) #thickness = -1 olunca içi dolu oluyor.
plt.imshow(blank_img)
```

```python
cv2.line(img=blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img,text='SELAM',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img) #org sol asağısını belli eder. linetype bu kullanılıyor genelde
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32) #her köşe için konum veriyorum x,y diye
```

```python
vertices.shape
```

```python
pts= vertices.reshape((-1,1,2)) # (-1,1,2) shape hep var.  3 boyuta çıkarmak için
```

```python
vertices.shape
```

```python
pts.shape # 3. boyut
```

```python
pts
```

```python
cv2.polylines(blank_img,pts=[pts],isClosed=True,color=(255,0,0),thickness=3)
plt.imshow(blank_img)
```
