```python
# üst ve alt thresholdlar seçeceğiz. Ona göre manuel ayarlama yapacağız
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("aykut full.jpg")
plt.imshow(img)
```

```python
edges= cv2.Canny(image=img,threshold1=0,threshold2=174)
plt.imshow(edges)
```

```python
#daha temiz algılamak ve noiseleri susturmak için
med_val = np.median(img)
med_val
```

```python
lower = int(max(0,0.7*med_val)) #alt sınırı ya 0 ya da median değerin %70i yaptı, hangisi büyükse)
upper= int(max(255,1.3*med_val)) #üst sınır ya median değeri %130 u ya da max value, hangisi küçükse
```

```python
edges= cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges) #bizim image'da daha kötü işledi,
#blur vs atarak arkada noise varsa onları yok edebiliriz.
```
