```python
'''Gradyant= verinin yönlü değişimi, soldan sağa beyazlama gibi. Böyle değişimleri algılyayan algoritmalar var.
Bu kurs için sobel-feldman operator kullanılacak.
x gradient kullanılırsa y eksenindeki dikeyler çıkar.
y gradient için tam tersi olur.
normalized gradient magnitude ise her ikisini alır.'''
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img): #bu fonksiyon çok kullanılıyor. büyük gözükmesi için figsize var
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')    
```

```python
img=cv2.imread('image6.jpg',0)
display_img(img)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5,) #1, x ekseni için
display_img(sobelx)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5,) #1, y ekseni için
display_img(sobely)
```

```python
#laplace operatörü var 2 boyutlu hali. x ve y yapıyor
laplacian= cv2.Laplacian(img,cv2.CV_64F)
display_img(laplacian)
```

```python
blended = cv2.addWeighted(src1=sobelx,alpha=0.7,src2=sobely,beta=0.3,gamma=0)
display_img(blended)
```
