```python
#Harris corner ve shi tomasi sistemleri.
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('image10.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess=cv2.imread('image9.jpeg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray = np.float32(gray_flat_chess) #floata çeviriyor.
```

```python
gray
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()]= [255,0,0] #rgb, max değer
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04) #default değerler
dst = cv2.dilate(dst,None)
real_chess[dst>0.01*dst.max()]= [255,0,0] #rgb, max değer
```

```python
plt.imshow(real_chess)
```

```python
flat_chess = cv2.imread('image10.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chess=cv2.imread('image9.jpeg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,maxCorners=100,qualityLevel=0.01,minDistance=10) #max corner sayısı verebiliyoruz. shi thomasi burada bunlar de default değer.
```

```python
corners
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),4,(255,0,0),-1) #diziyi yuvarlayıp circle çizmek için
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,maxCorners=100,qualityLevel=0.01,minDistance=10) #max corner sayısı verebiliyoruz. shi thomasi burada bunlar de default değer.
corners=np.int0(corners) #çizebilmek için int yapıyoruz.
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),4,(255,0,0),-1) #circle çizmek için

plt.imshow(real_chess)    
```
