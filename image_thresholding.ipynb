```python
img=cv2.imread('image4.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img): #bu fonksiyon ok kullanlyor. byk gzkmesi iin figsize var
    fig = plt.figure(figsize=(10,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret,th1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1) #bir snr deer belirledik ve ona gre binary uyguladk.
```

```python
#adaptive threshold uyguladk buna
th2=cv2.adaptiveThreshold(img,maxValue=255,adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C,thresholdType=cv2.THRESH_BINARY,blockSize=11,C=6)
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1,0.4,th2,0.6,0) #ikisini blendledik.
show_pic(blended)
```
