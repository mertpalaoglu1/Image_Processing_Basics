```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(10,10))
    ax= fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
#OKU VE GSTER
img = cv2.imread('image8.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#BINARY THRESHOLD uygula
img = cv2.imread('image8.webp',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap='gray')
```

```python
#hsv ile ama
img =cv2.imread('image8.webp')
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
# 4x4 kernelli lowpass filter oluturma ve 0.01 verileri ile 2D CONV. yapma !!!!
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
kernel
```

```python
img =cv2.imread('image8.webp')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
results = cv2.filter2D(img,-1,kernel=kernel)
plt.imshow(results)
```

```python
#horizontal sobel filter
img =cv2.imread('image8.webp',0)
sobelx= cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) # y=0 olacak.
plt.imshow(sobelx,cmap='gray')
```

```python
# r g b histogram oluturma
img =cv2.imread('image8.webp')

color = ['b','g','r'] # list atyoruz.
for i, col in enumerate(color): #enumare tekrar atyor.
    histr= cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
plt.title('histogram')
```
