```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('image1.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV) #HUE SATURATION VALUE, 
plt.imshow(img) 
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HLS) #HUE light saturation, 
plt.imshow(img) 
```
