```python
# FARKLI BOYUTLARDA OLURSA:
#OVERLAY SMALL IMAGE ON TOP OF A LARGERE IMAGE(NO BLENDING) [üst üste koyduk hep]
#NUMPY reassignment
```

```python
img1 = cv2.imread('image1.jpg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2= cv2.imread('image2.jpg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2=cv2.resize(img2,(100,100))
```

```python
large_img=img1
small_img=img2
```

```python
x_offset=0
y_offset=0 #kesme için sol üstten başlıyoruz.
```

```python
x_end= x_offset + small_img.shape[1] # xteki uzunluğu shape için tupple şeklinde ters geliyor. 
y_end= y_offset + small_img.shape[0] #ydeki uzunluğu, offsetler 0 olduğu için bir manası yok burada en soldan eklediğimiz için.
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img #o noktalardaan başlayarak diğer şeklinde üzerine kyodum şekli.
```

```python
plt.imshow(large_img)
```

```python
## MASKELEMEYİ YAPMA
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv=cv2.bitwise_not(img2gray) #üstteki image'ın tam inversini alacak ki maske oluşssun
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
mask_inv.shape #renk kanalları gitti
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape,255,dtype=np.uint8) #renk kanalını ekledik.
```

```python
white_background.shape
```

```python
bk=cv2.bitwise_or(white_background,white_background,mask=mask_inv)
```
