```python
img1 = cv2.imread('image1.jpg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2= cv2.imread('image2.jpg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2=cv2.resize(img2,(100,100))
```

```python
large_img=img1
small_img=img2
```

```python
x_offset=0
y_offset=0 #kesme iin sol stten balyoruz.
```

```python
x_end= x_offset + small_img.shape[1] # xteki uzunluu shape iin tupple eklinde ters geliyor. 
y_end= y_offset + small_img.shape[0] #ydeki uzunluu, offsetler 0 olduu iin bir manas yok burada en soldan eklediimiz iin.
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img #o noktalardaan balayarak dier eklinde zerine kyodum ekli.
```

```python
plt.imshow(large_img)
```

```python
## MASKELEMEY YAPMA
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv=cv2.bitwise_not(img2gray) #stteki image'n tam inversini alacak ki maske olussun
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
mask_inv.shape #renk kanallar gitti
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape,255,dtype=np.uint8) #renk kanaln ekledik.
```
