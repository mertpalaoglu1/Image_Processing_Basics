```python
#Hatırlatmalar olacaktır, 10.02.25 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('image2.jpg') #dosyanın aynı folderda olması lazım yoksa oç okumuyor bir dıştan
```

```python
plt.imshow(img)
```

```python
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
new_img= img_rgb.copy()
```

```python
new_img=cv2.flip(new_img,1)
plt.imshow(new_img) #ters çevirme işi
```

```python
#red rectangle çizecek kod 
pt1=(75,25)
pt2=(125,75)
```

```python
cv2.rectangle(img_rgb,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=1)
plt.imshow(img_rgb)
```

```python
#cv2 ile daire çizme hatırlatma
def create_circle(event,x,y,flags,param):
    if event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),thickness=5)

img = cv2.imread('image2.jpg')
cv2.namedWindow(winname='bruce') #winnamele setback name aynı olacak
cv2.setMouseCallback('bruce',create_circle)    
while True:
    cv2.imshow('bruce',img)
    if cv2.waitKey(20) &   0xFF==27:
        break
cv2.destroyAllWindows()      
```
