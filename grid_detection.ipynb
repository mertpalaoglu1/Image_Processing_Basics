gridler kamerayı kalibre etmek için ve obje detectionda çok kullanılır. Opencvde gömülü olanlar var.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('satranc_tahtasi.jpg')
plt.imshow(flat_chess)
```

```python
found,corners=cv2.findChessboardCorners(flat_chess,(6,6)) #bulup bulmadığını kontrol için found
cv2.findCirclesGrid()# bu da var.
```

```python
found
```

```python
cv2.drawChessboardCorners(flat_chess,(3,3),corners=corners,found=True) #bulunca grid çizebiliyor burayaa ama bu image'i okumayamadı
```
