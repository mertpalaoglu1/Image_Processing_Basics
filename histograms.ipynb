```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
dark_horse = cv2.imread('image7.jpg') #Original BGR
show_horse=cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) #RGB TO SHOWIMAGE

rainbow = cv2.imread('image3.jpg')
show_rainbow=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

bricks = cv2.imread('image5.webp')
show_bricks=cv2.cvtColor(bricks,cv2.COLOR_BGR2RGB) 
#bunlardan birinde hesaplama yapacağız. BGR olanlar ile, diğeri ile göstereceğiz. 
```

```python
#OPENCV BGR şuan blue'yu aldık. 
hist_values = cv2.calcHist([bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
img = dark_horse
color = ('b','g','r') 
```

```python
for i,col in enumerate(color):
    histr= cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.title('At Histogrami')    
```
